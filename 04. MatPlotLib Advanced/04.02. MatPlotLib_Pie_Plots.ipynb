# MATPLOTLIB: PIE CHARTS

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

1. Create a pie chart for job_work_from_home column 

```python
df['job_work_from_home'].value_counts().plot(kind='pie')
plt.title('Work from Home Status')
plt.ylabel('')
```

2. Create a pie chart for job_title_short column 

```python
df['job_title_short'].value_counts().plot(kind='pie')
plt.title('Job Title')
plt.ylabel('')
```

--> hard to compare, hard to know which one is higher than the others

3. Reformat the pie chart of work from home status

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle = 90, autopct='%1.2f%%')
# use startangle param to change the starting angle of the pie
plt.title('Work from Home Status')
plt.ylabel('')
```

- Common string codes for pie chart's data label:

'%1f' = 8.8523

'%1.1f' = 8.8 -> .1 is one decimal place, .2 is two decimal place

'%1.1f%%' = 8.8% -> double percent sign %% will add a percent sign after the data label 

3. Create 3 pies chart together using subplots

```python
df[['job_work_from_home','job_no_degree_mention','job_health_insurance']]
```

```python
fig, ax = plt.subplots(1,3)

dict_column = {
    'job_work_from_home':'Work From Home',
    'job_no_degree_mention':'Job Degree Req.',
    'job_health_insurance':'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()): # enumerate turns key-value pairs into tuples
    ax[i].pie(df[column].value_counts(),startangle=90, autopct= '%1.1f%%',labels=['False','True'])
    ax[i].set_title(title)
```

4. Use Pandas Plotting functions instead

```python
fig, ax = plt.subplots(1,3)

dict_column = {
    'job_work_from_home':'Work From Home',
    'job_no_degree_mention':'Job Degree Req.',
    'job_health_insurance':'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()): # enumerate turns key-value pairs into tuples
    df[column].value_counts().plot(kind='pie', autopct = '%1.1f%%', startangle = 90 ,ax = ax[i])
    ax[i].set_ylabel('')
    ax[i].set_title(title)
```
