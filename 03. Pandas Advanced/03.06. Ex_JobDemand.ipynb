# EXERCISE: MONTHLY JOB DEMAND

```python
# Import Libs
import matplotlib as mpl
import pandas as pd
from datasets import load_dataset
# Loading Data
dataset = load_dataset(path='lukebarousse/data_jobs')
df = dataset["train"].to_pandas()

# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df_VN = df[df['job_country'] == 'Vietnam'].copy()
df_VN
```

## create monthname column

- Different codes can be input as an string argument in the 'dt.strftime' function
- for example, "%B" will return the verbiage words for the months (January,..)

```python
df_VN['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')
df_VN
```

```python
df_VN_pivot = df_VN.pivot_table(index='job_posted_month',columns= 'job_title_short',aggfunc='size')
```

```python
df_VN_pivot.reset_index(inplace=True)
df_VN_pivot['month_no'] = pd.to_datetime(df_VN_pivot['job_posted_month'],format= '%B').dt.month

df_VN_pivot.sort_values(by= 'month_no',inplace=True)
df_VN_pivot.set_index(keys= 'job_posted_month',inplace= True)
df_VN_pivot.drop(columns= 'month_no',inplace=True)
```

```python
df_VN_pivot
```

## It's plotting time babyyyy

```python
df_VN_pivot.plot(kind='line')
```

- get the top 3 countries with most job openings in VN

```python
top_3 = df_VN['job_title_short'].value_counts().sort_values().head(3).index
df_VN_pivot = df_VN_pivot[top_3]
df_VN_pivot.plot(kind='line')
```
