# INDEX MANAGEMENT

```python
# Import Libs
import matplotlib as mpl
import pandas as pd
from datasets import load_dataset
# Loading Data
dataset = load_dataset(path='lukebarousse/data_jobs')
df = dataset["train"].to_pandas()

# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)

df.sample(5)
```

1. Index property of the Pandas DataFrame

```python
df.index
```

2. Index.name property

- this property is just a value holder, in this case, it holds the name of the index column (the leftmost column of the DF)

```python
df.index.name = 'job_index'

df.head()
```

3. index.dtype Property

```python
df.index.dtype # return the data type of the index
```

4. reset_index() Function

```python
df_usa = df[df['job_country'] == "United States"]
df_usa = df_usa.head(10)
df_usa
```

--> the index of the df_usa DataFrame is not consecutive, how to fix that??? use fucntion reset_index, and remember set the inplace param equal TRUE to modify the DataFrame only, not create a new Dataframe

```python
df_usa.reset_index(inplace=True)
df_usa
```

--> in case there is a need to merge data from other tables, the old index would be of much help

5. Set_index() function

```python

df_usa.set_index(keys='job_index',inplace = True)
df_usa
```

--> this function wil just drop the 'former' index column of the table

--> once you already set a column as an index column, you can't do that again right after that.

6. sort_index Function()

- This function will be much useful when we are grouping or pivoting the data

```python
median_pvt = df.pivot_table(values='salary_year_avg', index = 'job_title_short',aggfunc= ['median','min','max'])
median_pvt
```

```python
median_pvt.sort_values(by= ('median','salary_year_avg'),inplace= True, ascending= False)
median_pvt 
```

--> Because there are 2 levels of column titles (median and salary_year_avg) so if we want to sort by median of the salary_year_avg, we will have to feed the **by** param with a tuple containing those two column titles

```python
median_pvt.sort_index()
```

--> sort_index fx sort the table by the 1st column which is "job_title_short", and because this column stores sting type, it sorts in the alphabetical order.
