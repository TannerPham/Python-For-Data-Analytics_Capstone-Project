# Exercise: Investigate Trending Skills

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

1. Keep only "Data Analyst" job postings

```python
df_DA = df[df['job_title_short']== 'Data Analyst'].copy()

df_DA
```

2. Add a month column and use pivot tables to group the dataset 

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

df_DA_explode = df_DA.explode('job_skills')

df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_no',columns='job_skills',aggfunc= 'size', fill_value= 0) # dont know how to fill null values with 0, check this out

df_DA_pivot.plot(kind='line')
```

--> Oops, legend never dies, has to find a way to narrow down on the categories

3. Create a Grand Total row for the PV

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

df_DA_pivot
```

4. Get a list of skills which is in the descending order by number of counts

```python
df_DA_pivot.loc['Total'].sort_values(ascending=False).index
```

5. Apply this list to the df_DA_pivot to sort the columns accordingly

```python
df_DA_pivot= df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
```

6. Drop the Grand Total row

```python
df_DA_pivot = df_DA_pivot.drop('Total')
df_DA_pivot
```

7. Let's plot itttt babyyyy

```python
df_DA_pivot.iloc[:,:5].plot(kind='line')

plt.title('Top 5 skills for DA per Month')
plt.xlabel('')
plt.ylabel('Job Posting Counts')
```

8. Let's change the month num to the month name babyyy

```python
# change from month number to month name
df_DA_pivot = df_DA_pivot.reset_index()
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format= '%m').strftime('%b'))
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')


# plot it again
df_DA_pivot.iloc[:,:5].plot(kind='line')

plt.title('Top 5 skills for DA per Month')
plt.xlabel('')
plt.ylabel('Job Posting Counts')
```
